# ----------------------------------------------------------------------
# STEP 2: Data Type Conversion
# Documentation: Converting 'InvoiceDate' to datetime format for time-series analysis
# and converting 'CustomerID' to object/string since it's an identifier, not a numeric scale.
# ----------------------------------------------------------------------

```python

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['CustomerID'] = df['CustomerID'].astype(str).str.split('.').str[0] 
```

# ----------------------------------------------------------------------
# STEP 3: Handling Missing Values
# Documentation: 
# - 'Description' missing is minor, but 'CustomerID' missing is critical for 
#   Customer Segmentation. Transactions without CustomerID cannot be grouped 
#   into individual customer profiles, so we must drop them.
# ----------------------------------------------------------------------
# Drop rows where CustomerID is missing (originally NaN, now 'nan' string after conversion)

```python
df = df[df['CustomerID'] != 'nan']
df = df.dropna(subset=['Description'])
```

# ----------------------------------------------------------------------
# STEP 4: Handling Duplicates
# Documentation: Checking for and removing exact duplicate transactions to avoid 
# double-counting sales.
# ----------------------------------------------------------------------

```python
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows found and removed: {duplicate_count}")
df = df.drop_duplicates()
```

# ----------------------------------------------------------------------
# STEP 5: Handling Cancellations and Negative Values
# Documentation: 
# - Transactions with negative 'Quantity' or 'UnitPrice' <= 0 are faulty or cancellations.
# - Invoices starting with 'C' are cancellations. We will separate them into a different 
#   dataframe if needed for returns analysis, and keep only valid successful purchases (Quantity > 0)
#   and valid prices (UnitPrice > 0) for the standard RFM metrics.
# ----------------------------------------------------------------------
# Separate Cancellations for reference

```python
cancellations = df[df['InvoiceNo'].str.startswith('C',na=False) | (df['Quantity'] < 0)]
cancellations
```

# Keep only positive quantities and prices for pure sales analysis

```python
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

# ----------------------------------------------------------------------
# STEP 6: Feature Engineering (Revenue & Time Features)
# Documentation: Creating 'TotalRevenue' (Quantity * UnitPrice) and extracting 
# temporal components (Month, Day of Week, Hour) for exploratory analysis.
# ----------------------------------------------------------------------

```python
df['TotalRevenue'] = df['Quantity'] * df['UnitPrice']
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')
df['InvoiceMonthStr'] = df['InvoiceDate'].dt.strftime('%B')
df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()
df['Hour'] = df['InvoiceDate'].dt.hour

print("Cleaned Dataset Shape:", df.shape)
df.head()
```

# Customer Segmentation

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
```

# ----------------------------------------------------------------------
# STEP 1: Engineer RFM Features
# Documentation: Calculate Recency, Frequency, and Monetary for each customer.
# ----------------------------------------------------------------------
# We assume the default day date is one day after the last invoice to calculate the differences.

```python
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days, 
    'InvoiceNo': 'nunique',                                  
    'TotalRevenue': 'sum'                                    
}).rename(columns={'InvoiceDate': 'Recency', 'InvoiceNo': 'Frequency', 'TotalRevenue': 'Monetary'})

print("RFM Features Sample:")
print(rfm.head())
```

# ----------------------------------------------------------------------
# STEP 2: Scale Data Appropriately
# Documentation: Standardize the features to have a mean of 0 and variance of 1.
# ----------------------------------------------------------------------

```python
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)
```

# ----------------------------------------------------------------------
# STEP 3 & 4: Evaluation Methods (Elbow Method)
# Documentation: Run K-Means for a range of clusters to find the "Elbow".
# ----------------------------------------------------------------------

```python
inertia = []
k_range = range(2, 9)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(rfm_scaled)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(k_range, inertia, marker='o', color='purple', linewidth=2)
plt.title('The Elbow Method to Find Optimal K', fontsize=14)
plt.xlabel('Number of Clusters (K)', fontsize=12)
plt.ylabel('Inertia / WCSS', fontsize=12)
plt.grid(True)
plt.show()
```

```python

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)


rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)


cluster_analysis = rfm.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'Cluster': 'count'  
}).rename(columns={'Cluster': 'CustomerCount'}).reset_index()

print("--- Clusters Profile (Mean Values) ---")
print(cluster_analysis)
```

```python
cluster_map = {
    2: "VIP / Champions",
    3: "Loyal Customers",
    0: "Active / General",
    1: "Churned / Lost"
}
rfm['Segment'] = rfm['Cluster'].map(cluster_map)


plt.figure(figsize=(11, 6))


ax = sns.countplot(
    data=rfm, 
    x='Segment', 
    order=["VIP / Champions", "Loyal Customers", "Active / General", "Churned / Lost"], 
    palette='Set2'
)

for container in ax.containers:
    ax.bar_label(container, fontsize=12, weight='bold', padding=3)


plt.title('Customer Distribution by Segment', fontsize=16, pad=15, weight='bold')
plt.xlabel('Customer Segment', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)

plt.tight_layout()
plt.show()
```

## Customer Segmentation using RFM & K-Means Clustering

# 1. Feature Engineering: RFM Analysis


Description: > To analyze and segment customer purchasing behavior, we engineer three core behavioral metrics from the transactional data, collectively known as RFM:

Recency ($R$): The number of days since the customer's last purchase. A lower recency indicates higher engagement.
Frequency ($F$): The total number of unique orders placed by the customer. It measures loyalty and buying consistency.
Monetary ($M$): The total financial value (spending) generated by the customer. It reflects the customer's direct revenue impact.

# 2. Data Preprocessing: Standardization
Description: > Distance-based algorithms like K-Means are highly sensitive to the scale of the input features. Since Monetary values are in thousands, while Recency values span days, we apply StandardScaler to standardize the RFM vectors to a mean of $0$ and a variance of $1$. This ensures all three features contribute equally to the cluster assignments.


# 3. Algorithm Selection & Optimization (The Elbow Method)
Description: > We chose K-Means Clustering due to its computational efficiency on transactional business datasets and its direct interpretability for marketing actions.

To determine the optimal number of clusters ($K$), we evaluated the Within-Cluster Sum of Squares (WCSS / Inertia) across a range of $K$ values ($2$ to $8$).

The resulting Elbow Curve showed a distinct inflection point (the "Elbow") at $K = 4$, indicating the optimal balance between minimizing intra-cluster variance and avoiding over-segmentation.

The hyperparameter n_init=10 was utilized to run the algorithm with $10$ different random centroid initializations, ensuring the stability and convergence of the final global optimum.


# 4. Business Interpretation of Customer Segments

Based on the computed cluster centroids, the customer base is divided into four strategically distinct behavioral archetypes:

1. VIP / Champions (Cluster 2): * Behavioral Profile: Extremely low recency (active within the last week), exceptionally high order frequency, and massive total expenditure.


Business Action: Retain at all costs through exclusive loyalty benefits, priority customer service, and direct relationship management.

2. Loyal Customers (Cluster 3): * Behavioral Profile: Low recency, high frequency, and solid financial spending.


Business Action: Implement up-selling and cross-selling marketing campaigns for new arrivals, as they exhibit high trust in the brand.

3. Active / General Market (Cluster 0): * Behavioral Profile: Moderate recency, stable but lower order frequency, and average spending. This represents the absolute majority (mass volume) of the company's customer base.


Business Action: Engage with periodic seasonal promotions, discount codes, and personalized newsletters to increase their purchase frequency.

4. Churned / Lost Customers (Cluster 1): * Behavioral Profile: Very high recency (inactive for over 8 months), minimal frequency (single purchases), and low lifetime value.


Business Action: Avoid expensive broad marketing spending. Execute a targeted win-back/re-activation campaign with heavy discounts as a final attempt to regain them.
